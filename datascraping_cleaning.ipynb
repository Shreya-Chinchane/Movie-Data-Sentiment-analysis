# Extract reviews for all movies

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import pandas as pd
```

```python
driver = webdriver.Chrome()
```

```python
driver.get('https://www.imdb.com/list/ls564506313/?user_rating=4%2C7.9')
```

```python
import re

# Find all movie links
movie_links = driver.find_elements(By.CSS_SELECTOR, 'a.ipc-lockup-overlay.ipc-focusable')

# Print all hrefs to verify
for link in movie_links:
    href = link.get_attribute("href")
    print(href)  # Print to check the href format

# List to store extracted IMDb IDs
imdb_ids = []

# Regular expression to extract the 'tt' ID
pattern = r"/title/(tt\d+)/"

# Loop through each link, extract href, and apply regex
for link in movie_links:
    href = link.get_attribute("href")
    match = re.search(pattern, href)
    if match:
        imdb_ids.append(match.group(1))
```

```python
# Base URL for reviews
base_url = 'https://www.imdb.com/title/{}/reviews'

# Dictionary to store reviews for each movie ID
all_reviews = {}

# Loop through each IMDb ID and scrape reviews
for movie_id in imdb_ids:
    try:
        # Construct the review URL
        url = base_url.format(movie_id)
        print(f"Fetching reviews for {movie_id}...")

        # Use the same driver to navigate to the review page
        driver.get(url)
        time.sleep(2)  # Allow time for the page to load

        # Extract review titles
        review_elements = driver.find_elements(By.CSS_SELECTOR, 'h3.ipc-title__text')
        reviews = [review.text for review in review_elements if review.text.strip()]

        # Store reviews in a dictionary
        all_reviews[movie_id] = reviews

        print(f"Found {len(reviews)} reviews for {movie_id}")
    except Exception as e:
        print(f"Error fetching reviews for {movie_id}: {e}")
```

```python
# Convert the results to a DataFrame
df = pd.DataFrame({'Movie_ID': list(all_reviews.keys()), 'Reviews': list(all_reviews.values())})
```

```python
# Save to a CSV file (optional)
df.to_excel('moderate_reviews.xlsx', index=False)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import time
```

```python
# Construct URLs dynamically with the required format
urls = [f"https://www.imdb.com/title/{imdb_id}/?ref_=chttp_i_{index+1}" for index, imdb_id in enumerate(imdb_ids)]

# Create an empty list to store movie data
movies_data = []

# Loop through each URL
for url in urls:
    driver.get(url)  # Load the page
    time.sleep(2)  # Allow time for the page to load fully

    # Extract IMDb ID from the URL
    imdb_id = url.split('/')[4]  # 'tt0111161' is the 5th element in the split result

    # Extract the genres
    try:
        genres_element = driver.find_element(By.CLASS_NAME, 'ipc-chip-list--baseAlt')
        genres = genres_element.text.splitlines()
    except Exception:
        genres = []

    # Extract the movie description
    try:
        about = driver.find_element(By.CLASS_NAME, 'sc-3ac15c8d-3').text
    except Exception:
        about = ""

    # Extract the rating
    try:
        rating = driver.find_element(By.CLASS_NAME, 'sc-d541859f-1').text
    except Exception:
        rating = ""

    # Extract the rating count
    try:
        rating_count = driver.find_element(By.CLASS_NAME, 'sc-d541859f-3').text
    except Exception:
        rating_count = ""

    # Extract the director's name
    try:
        director = driver.find_element(By.CLASS_NAME, 'ipc-metadata-list-item__list-content-item').text
    except Exception:
        director = ""

    # Extract the writers
    try:
        writers_elements = driver.find_elements(By.CLASS_NAME, 'ipc-metadata-list-item__content-container')
        if len(writers_elements) > 1:
            writers_raw = writers_elements[1].text
            # Fixing the issue of concatenated names
            writers = ' '.join(writers_raw.split())
        else:
            writers = ""
    except Exception:
        writers = ""

    # Append the extracted data to the list
    movies_data.append({
        'IMDb ID': imdb_id,
        'Genres': genres,
        'Description': about,
        'Rating': rating,
        'Rating Count': rating_count,
        'Director': director,
        'Writers': writers
    })
# Create a DataFrame
df = pd.DataFrame(movies_data)

# Display the DataFrame
print(df)
```

```python

from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import time


# Construct URLs dynamically with the required format
urls = [f"https://www.imdb.com/title/{imdb_id}/?ref_=chttp_i_{index+1}" for index, imdb_id in enumerate(imdb_ids)]

# Initialize the WebDriver
driver = webdriver.Chrome()

# Create an empty list to store movie data
movies_data = []

# Loop through each URL
for url in urls:
    driver.get(url)  # Load the page
    time.sleep(2)  # Allow time for the page to load fully

    # Extract IMDb ID from the URL
    imdb_id = url.split('/')[4]  # 'tt0111161' is the 5th element in the split result

    # Extract the movie description
    try:
        about = driver.find_element(By.CLASS_NAME, 'sc-3ac15c8d-3').text
    except Exception:
        about = ""

    # Extract the rating
    try:
        rating = driver.find_element(By.CLASS_NAME, 'sc-d541859f-1').text
    except Exception:
        rating = ""

    # Extract the rating count
    try:
        rating_count = driver.find_element(By.CLASS_NAME, 'sc-d541859f-3').text
    except Exception:
        rating_count = ""

    # Extract the director's name
    try:
        director = driver.find_element(By.CLASS_NAME, 'ipc-metadata-list-item__list-content-item').text
    except Exception:
        director = ""

    # Extract the writers
    try:
        writers = ""
        writers_elements = driver.find_elements(By.CLASS_NAME, 'ipc-metadata-list-item__content-container')
        if len(writers_elements) > 1:
            writers_container = writers_elements[1]
            writer_names = writers_container.find_elements(By.TAG_NAME, 'a')  # Extract names from <a> tags
            writers = ", ".join([writer.text for writer in writer_names])
    except Exception:
        writers = ""

    # Append the extracted data to the list
    movies_data.append({
        'IMDb ID': imdb_id,
        'Description': about,
        'Rating': rating,
        'Rating Count': rating_count,
        'Director': director,
        'Writers': writers
    })


# Create a DataFrame
df = pd.DataFrame(movies_data)

# Display the DataFrame
print(df)

```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import time


# Construct URLs dynamically
urls = [f"https://www.imdb.com/title/{imdb_id}/?ref_=chttp_i_{index+1}" for index, imdb_id in enumerate(imdb_ids)]


# Create an empty list to store movie data
movies_data = []

# Loop through each URL
for url in urls:
    driver.get(url)  # Load the page
    time.sleep(2)  # Allow time for the page to load fully

    # Extract IMDb ID
    imdb_id = url.split('/')[4]

    # Extract stars
    try:
        stars_element = driver.find_elements(By.CSS_SELECTOR, 'ul.ipc-inline-list')
        if len(stars_element) > 4:  # Ensure the 5th element exists
            raw_stars_text = stars_element[4].text.strip()
            # Split the raw text into separate names and join them with commas
            stars = ", ".join(raw_stars_text.splitlines())
        else:
            stars = ""
    except Exception:
        stars = ""

    # Append the extracted data to the list
    movies_data.append({
        'IMDb ID': imdb_id,
        'Stars': stars
    })


# Create a DataFrame from the extracted data
df = pd.DataFrame(movies_data)

# Display the DataFrame
print(df)

```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import time

# Base URL for reviews
base_url = 'https://www.imdb.com/title/{}/reviews'


# Dictionary to store reviews and titles for each movie ID
all_reviews = {}

# Loop through each IMDb ID and scrape reviews and titles
for movie_id in imdb_ids:
    try:
        # Construct the review URL
        url = base_url.format(movie_id)
        print(f"Fetching reviews for {movie_id}...")

        # Navigate to the review page
        driver.get(url)
        time.sleep(2)  # Allow time for the page to load

        # Extract the movie title
        try:
            title_element = driver.find_element(By.XPATH, '//h2[@data-testid="subtitle"]')
            movie_title = title_element.text.strip()
        except Exception as e:
            print(f"Error fetching title for {movie_id}: {e}")
            movie_title = "N/A"

        # Extract review titles
        try:
            review_elements = driver.find_elements(By.CSS_SELECTOR, 'h3.ipc-title__text')
            reviews = [review.text for review in review_elements if review.text.strip()]
        except Exception as e:
            print(f"Error fetching reviews for {movie_id}: {e}")
            reviews = []

        # Store reviews and titles in a dictionary
        all_reviews[movie_id] = {'Movie Title': movie_title, 'Reviews': reviews}

        print(f"Found {len(reviews)} reviews for '{movie_title}' ({movie_id})")
    except Exception as e:
        print(f"Error processing {movie_id}: {e}")


# Convert the results to a DataFrame
data = {
    'Movie_ID': [],
    'Movie Title': [],
    'Reviews': []
}

for movie_id, details in all_reviews.items():
    data['Movie_ID'].append(movie_id)
    data['Movie Title'].append(details['Movie Title'])
    data['Reviews'].append(details['Reviews'])

df = pd.DataFrame(data)

# Display the DataFrame
print(df)


```

```python
# Optionally save to CSV
df.to_excel('moderate_reviews.xlsx', index=False)
```

```python
import pandas as pd
import numpy as np

# Define country-to-language mapping
country_to_language = {
    'United States': 'English',
    'United Kingdom': 'English',
    'New Zealand': 'English',
    'India': 'Hindi',
    'France': 'French',
    'Germany': 'German',
    'Italy': 'Italian',
    'Japan': 'Japanese',
    'Brazil': 'Portuguese',
    'South Korea': 'Korean',
    'Spain': 'Spanish',
    'Argentina': 'Spanish',
    'Canada': 'English',  
    'Australia': 'English',
    'Denmark': 'Danish',
    'Iran': 'Persian',
    'Turkey': 'Turkish',
    'Sweden': 'Swedish',
    'Lebanon': 'Arabic',
    'Mexico': 'Spanish',
    'West Germany': 'German',
    'Soviet Union': 'Russian', 
    'Ireland': 'English'
}

# Fill missing languages by looking up the country
top['Language'] = top.apply(
    lambda row: country_to_language[row['Country']] if pd.isnull(row['Language']) else row['Language'], axis=1
)

```

```python
fixed_genres = [
    'Action', 'Adventure', 'Animation', 'Biography', 'Comedy', 'Crime', 'Drama', 'Family', 
    'Fantasy', 'History', 'Horror', 'Music', 'Mystery', 'Romance', 'Sci-Fi', 'Sport', 
    'Thriller', 'War', 'Western'
]

# Function to clean and map genres
def map_genres(genre_list, fixed_genres):
    # Convert the string representation of the list into an actual list
    genre_list = eval(genre_list)  # Be careful when using eval
    # Map to fixed genres
    unique_genres = [genre for genre in genre_list if genre in fixed_genres]
    return ', '.join(unique_genres)

# Apply the mapping function to create a new column
bottom['Unique Genre'] = bottom['Genres'].apply(lambda x: map_genres(x, fixed_genres))
```
